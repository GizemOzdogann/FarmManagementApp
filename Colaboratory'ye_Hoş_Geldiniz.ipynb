**Label encoding** ile kategorik verileri algoritmaların işleyebileceği sayısal forma dönüştürebiliriz.

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```
for column in mushroom_data.columns:
  mushroom_data[column] = le.fit_transform(mushroom_data[column])
```

```
mushroom_data.head()
```

Örneğin 'class' sütunundaki iki farklı veriyi p = 1, e = 0 olacak şekilde eşleştirerek numerik olarak dönüşüm sağladı. Bu sayede üzerinde işlem yapılabilecek bir formata sahip oldu.

```
X = mushroom_data.drop(columns = "class")
```
